```python
import ee
import geemap

Map = geemap.Map()

image = ee.ImageCollection('LANDSAT/LC08/C02/T1_L2') \
    .filterDate('2022-01-01', '2022-12-31') \
    .median()

Map.addLayer(
    image,
    {
        'bands': ['SR_B4', 'SR_B3', 'SR_B2'],
        'min': 0,
        'max': 30000
    },
    'Satellite Image'
)

Map
```

```python
chennai = ee.Geometry.Point([80.2707, 13.0827])

Map.centerObject(chennai, 10)
```

```python
# OLD IMAGE (2015)
old = ee.ImageCollection('LANDSAT/LC08/C02/T1_L2') \
    .filterDate('2015-01-01', '2015-12-31') \
    .median()

# NEW IMAGE (2023)
new = ee.ImageCollection('LANDSAT/LC08/C02/T1_L2') \
    .filterDate('2023-01-01', '2023-12-31') \
    .median()
```

```python
old_ndvi = old.normalizedDifference(['SR_B5', 'SR_B4'])
new_ndvi = new.normalizedDifference(['SR_B5', 'SR_B4'])
```

```python
Map.addLayer(old_ndvi, {'min': -1, 'max': 1, 'palette': ['white', 'green']}, 'NDVI 2015')
Map.addLayer(new_ndvi, {'min': -1, 'max': 1, 'palette': ['white', 'green']}, 'NDVI 2023')
Map
```

```python
change = new_ndvi.subtract(old_ndvi)

Map.addLayer(change, {
    'min': -0.5,
    'max': 0.5,
    'palette': ['red', 'white', 'green']
}, 'Vegetation Change')

Map
```
